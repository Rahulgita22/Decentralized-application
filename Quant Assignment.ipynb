```python
import pandas as pd
import json
import ast # Required for safe evaluation of string literals

#Load the CSV files
rules_df = pd.read_csv('assignment_investment_rules.csv')
sector_df = pd.read_csv('assignment_sector_data.csv')
mcap_df = pd.read_csv('assignment_mcap.csv')

#Inspect first few rows
print(rules_df.head())
print(mcap_df.head())
print(sector_df.head())
```

```python
year_stock_dict = {}

for year in rules_df.columns[1:]:
    year_stock_dict[year] = []  # Initialize an empty list for the current year

    # Iterate over each row in the DataFrame
    for _, row in rules_df.iterrows():
        value = row[year]  # Extract the value for the current year and row

        # Check if the value is a string
        if isinstance(value, str):
            try:
                # Safely evaluate the string to a Python literal (e.g., list, dict)
                stock_list = ast.literal_eval(value)
            except (SyntaxError, ValueError):
                # Handle invalid literals by using the value as-is
                stock_list = value
        else:
            # If the value is not a string, use it as-is
            stock_list = value

        # Append the processed value to the year's list
        year_stock_dict[year].append(stock_list)

# Output the final dictionary
print(year_stock_dict)
```

```python
# Load the dataset containing company and sector data
df = pd.read_csv('assignment_sector_data.csv')

# Function to retrieve the sector for a given company name
def get_sector_for_company(company_name):
    """
    Get the sector corresponding to a given company name.

    Args:
        company_name (str): The name of the company.

    Returns:
        str: The sector name if the company is found; otherwise, "Unknown".
    """
    sector = df.loc[df['CO_NAME'] == company_name, 'Sector']
    # Return the sector if found, otherwise "Unknown"
    return sector.iloc[0] if not sector.empty else "Unknown"

# Create a list of all company names from the dataset
company_list = df['CO_NAME'].tolist()

# Build a dictionary mapping each company name to its sector
sectors = {
    company: get_sector_for_company(company) for company in company_list
}

# Print the resulting dictionary of sectors
print("Sectors:", sectors)
```

```python
import pandas as pd

def get_rank_range(file_name, start_rank, end_rank):
    """
    Retrieve a list of companies within the specified rank range for each year.

    Args:
        file_name (str): The name of the CSV file containing company data.
        start_rank (int): The start rank for sorting.
        end_rank (int): The end rank for sorting.

    Returns:
        dict: A dictionary with year as key and list of companies as value.
        str: Error message if input parameters are invalid or an exception occurs.
    """
    try:
        # Load the data from the provided CSV file
        df = pd.read_csv(file_name)

        # Validate the rank range input
        if start_rank < 1 or end_rank < start_rank:
            return "Invalid rank range. Please provide valid start and end ranks."

        # Create a dictionary to store the ranked companies for each year
        results = {}

        # Iterate over each year column (excluding 'CO_NAME')
        for year in df.columns[1:]:
            # Sort the companies based on the market capitalization (or similar metric for the year)
            sorted_df = df[['CO_NAME', year]].sort_values(by=year, ascending=False).reset_index(drop=True)

            # Extract the companies within the specified rank range
            ranked_companies = sorted_df.iloc[start_rank-1:end_rank]['CO_NAME'].tolist()
            results[year] = ranked_companies
        return results

    except FileNotFoundError:
        return f"File '{file_name}' not found. Please ensure the file exists in the specified path."
    except Exception as e:
        return f"An error occurred: {e}"

file_name = "assignment_mcap.csv"
start_rank = int(input("Enter the start rank: "))
end_rank = int(input("Enter the end rank: "))

# Retrieve the ranked companies based on the user input
ranked_stocks = get_rank_range(file_name, start_rank, end_rank)

# Output the result
print(ranked_stocks)
```

```python
def get_stocks_by_sector(sector_names):
    """
    Retrieve stocks by sector for each year, filtering by the specified sectors.

    Args:
        sector_names (list): A list of sector names to filter by.

    Returns:
        dict: A dictionary with year as the key and a list of matching stocks as the value.
    """
    matching_stocks_by_year = {}
    # Iterate over each year in the ranked stocks data
    for year, stocks in ranked_stocks.items():
        # Find stocks that match the specified sectors
        matching_stocks = [
            stock for stock in stocks if stock in sectors and sectors[stock] in sector_names
        ]
        # If there are any matching stocks, add them to the result dictionary
        if matching_stocks:
            matching_stocks_by_year[year] = matching_stocks
    return matching_stocks_by_year

# User input for sector names, separated by commas
user_sectors_str = input("Enter sector names (comma-separated): ").strip()

# Process the input into a list of sector names
user_sectors = [sector.strip() for sector in user_sectors_str.split(',')]

# Get the result for the specified sectors
result = get_stocks_by_sector(user_sectors)

# Output the result
if result:
    print(f"Stocks in the sectors {', '.join(user_sectors)} by year:")
    for year, stocks in result.items():
        print(f"{year}: {', '.join(stocks)}")
else:
    print(f"No stocks found in the sectors {', '.join(user_sectors)} for any year.")
```

```python
def filter_year_stock_dict(result, year_stock_dict):
    """
    Filters the stock lists by matching stocks between result and year_stock_dict for each year.

    Args:
        result (dict): A dictionary where each year maps to a list of valid stocks.
        year_stock_dict (dict): A dictionary where each year maps to lists/tuples of stocks.

    Returns:
        dict: A filtered dictionary where each year maps to a list of matching stocks.
    """
    filtered = {}
    # Iterate over each year in the year_stock_dict
    for year, stock_lists in year_stock_dict.items():
        # Check if the year exists in the result
        if year in result:
            # Filter stock lists based on the matching stocks in result[year]
            filtered_stock_lists = [
                [stock for stock in stock_list if stock in result[year]]
                for stock_list in stock_lists if isinstance(stock_list, (list, tuple))
            ]
            # Add non-empty lists to the filtered dictionary
            filtered[year] = [lst for lst in filtered_stock_lists if lst]
    return filtered

# Example usage of the function
filtered_dict = filter_year_stock_dict(result, year_stock_dict)

# Output the filtered dictionary
print(filtered_dict)
```

```python
import csv
from IPython.display import FileLink

# Strategy name for the output CSV file
strategy_name = "Updated_Strategy"

def write_dict_to_csv(strategy_name, filtered_dict, output_file):
    """
    Write the filtered stock data to a CSV file, including the strategy name.

    Args:
        strategy_name (str): The name of the strategy to be included in the first column.
        filtered_dict (dict): A dictionary where each year maps to a list of stocks.
        output_file (str): The file path where the CSV file will be saved.
    Returns:
        None
    """
    with open(output_file, mode="w", newline="") as file:
        writer = csv.writer(file)

        # Write the header row: 'Strategy Name' followed by the years
        header = ["Strategy Name"] + list(filtered_dict.keys())
        writer.writerow(header)

        # Determine the maximum number of rows (stock lists) across all years
        max_rows = max(len(stock_lists) for stock_lists in filtered_dict.values())

        # Write each row with the strategy name and the stocks for each year
        for i in range(max_rows):
            row = [strategy_name]  # Add strategy name in the first column
            for year in filtered_dict.keys():
                # Check if there are enough stocks for the current year
                if i < len(filtered_dict[year]):
                    row.append(", ".join(filtered_dict[year][i]))  # Join stocks with a comma
                else:
                    row.append("")  # Empty cell for years with fewer rows
            writer.writerow(row)

# Define output file name
output_file = "filtered_stocks.csv"

write_dict_to_csv(strategy_name, filtered_dict, output_file)

print(f"CSV file '{output_file}' has been created.")
FileLink(output_file)
```
